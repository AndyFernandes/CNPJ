# Leitura dos dados

```
path = "/content/drive/My Drive/CNPJ/data/dados_cnpj_fortaleza.csv"

data = pd.read_csv(path, encoding="ISO-8859-1")
print("Quantidade de dados: ", len(data))

data = data[['capital_social', 'cnae_fiscal', 'cnpj', 'cod_nat_juridica', 'matriz_filial', 'motivo_situacao', 'nome_fantasia', 'opc_simples', 'porte', 'qualif_resp', 'razao_social', 'situacao', 'telefone_1', 'telefone_2', 'data_inicio_ativ', 'data_situacao', 'email', 'tipo_logradouro', 'logradouro', 'numero', 'complemento', 'bairro' ]]
data.head()
```

# Gerando endereços

```
def format_address(data):
    enderecos = []
    for index, row in data.iterrows():
        address = ''
        if row['tipo_logradouro'] == row['tipo_logradouro']:
            address = str(row['tipo_logradouro'])
        if row['logradouro'] == row['logradouro']:
            address += ' ' + str(row['logradouro'])
        if row['numero'] == row['numero']:
            address += ', ' + str(row['numero']) 
        if row['bairro'] == row['bairro']:
            address += ', '  + str(row['bairro'])
        address += ', FORTALEZA, CEARA'

        enderecos.append(unidecode(address))
    return enderecos

data['endereco'] = format_address(data)
data['lat'] = [0 for _ in range(data.shape[0])]
data['lon'] = [0 for _ in range(data.shape[0])] 
```

# Limpando os nulos

```
data.fillna(0, inplace=True)
data.head()
```

# Convertendo tipos

```
data['cnae_fiscal'] = data['cnae_fiscal'].apply(lambda x: round(x))
data['cnpj'] = data['cnpj'].apply(lambda x: round(x))
data['cod_nat_juridica'] = data['cod_nat_juridica'].apply(lambda x: round(x))
data['data_inicio_ativ'] = data['data_inicio_ativ'].apply(lambda x: round(x))
data['data_situacao'] = data['data_situacao'].apply(lambda x: round(x))
data['matriz_filial'] = data['matriz_filial'].apply(lambda x: round(x))
data['motivo_situacao'] = data['motivo_situacao'].apply(lambda x: round(x))
data['opc_simples'] = data['opc_simples'].apply(lambda x: round(x))
data['porte'] = data['porte'].apply(lambda x: round(x))
data['qualif_resp'] = data['qualif_resp'].apply(lambda x: round(x))
data['situacao'] = data['situacao'].apply(lambda x: round(x))
```
